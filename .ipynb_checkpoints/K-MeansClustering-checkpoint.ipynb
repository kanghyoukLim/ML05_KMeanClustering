K-means clustering은 데이터를 입력받아 소수의 그룹으로 묶는 알고리즘  
레이블이 없는 데이터를 입력받아 각 데이터에 레이블을 할당해서 군집화를 수행  
  
  학습과정  
  1.중심점 선택 --> 2.클러스터 할당 --> 3.새로운 중심점 선택 --> 4.범위 확인  
  > 중심점 선택 : 랜덤하게 초기 중심점 선택  
  > 클러스터 할당 : K개 각 중심점에서 개별 데이터 간의 거리 측정하여 클러스터 형성
  > 새로운 중심점 선택 : 형성된 클러스터의 새로운 중심점을 계산  
  > 범위확인 : 새로운 중심점에서 개별 데이터 간의 거리를 재 측정하여 새로운 클러스터 형성, 새 클러스터가 형성되면 중심점을 다시 계산하고 클러스터 재 형성, 이 과정이 반복됨.  
  > 더이상 변화가 없다면 진행 멈춤

#### 적절한 K값을 찾는 것이 목표인 예제

분석절차   
> 라이브러리 호출 --> 데이터셋 로딩 --> 훈련과 검증 데이터셋 분리 --> 모델 생성 --> K값 예측

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('sales data.csv')
data.head()
```

```python
categorical_features = ['Channel', 'Region']
continuous_features = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']

for col in categorical_features:
    dummies = pd.get_dummies(data[col], prefix=col)
    data = pd.concat([data, dummies], axis=1)
    data.drop(col, axis=1, inplace=True)
data.head()
```

```python
mms = MinMaxScaler()
mms.fit(data)
data_transformed = mms.transform(data)
```

```python
mms = MinMaxScaler()
mms.fit(data)
data_transformed = mms.transform(data)

Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data_transformed)
    Sum_of_squared_distances.append(km.inertia_)

plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Optimal k')
plt.show()
```
